```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df
# to convert the date into datetime object if it is in string format 
```

```python
df['month'] = df['Date'].dt.month
df['day_of_week'] = df['Date'].dt.day_name()
df['year']=df['Date'].dt.year
# extracting date, month , year from the date column
```

```python
weekly_sales=df[['Holiday_Flag','Weekly_Sales']]
average_weekly_sales = weekly_sales.groupby('Holiday_Flag')['Weekly_Sales'].mean()
average_weekly_sales

# to check the average sales when there is holidays and when there is working day
```

```python



average_weekly_sales.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Holiday_Flag')
plt.ylabel('Average Weekly Sales')
plt.title('Average Weekly Sales by Store')
plt.show()
```

```python
weekly_sales=df[['Holiday_Flag','Weekly_Sales']]
total_weekly_sales = weekly_sales.groupby('Holiday_Flag')['Weekly_Sales'].sum()
average_weekly_sales
## to check the overall sales when there is holidays and when there is working day
```

```python

total_weekly_sales.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Holiday_Flag')
plt.ylabel('total_weekly_sales')
plt.title('Total Weekly Sales by Store')
plt.show()

```

```python
#Obseravtion
#From the above plot we can see that when there is a holiday the average weekly sales increase.
#However, over all total sales on holidays are lower due to the limited number of holidays each year, 
#which naturally reduces the overall total.
```

```python
average_unemployment_year = df.groupby('year').agg({'Unemployment': 'mean'})
average_unemployment_year
# to check the unemployment rate over the years
```

```python

plt.plot(average_unemployment_year.index, average_unemployment_year['Unemployment'], linestyle='-')

plt.xlabel('Year')
plt.ylabel('Average Unemployment')
plt.title('Trend of Average Unemployment over the Years')

plt.grid(True)
plt.show()

```

```python
average_sales_year = df.groupby('year').agg({'Weekly_Sales': 'mean'})
average_sales_year
# to check the average sales over the years to know the trend
```

```python
plt.plot(average_sales_year.index, average_sales_year['Weekly_Sales'], linestyle='-')

plt.xlabel('Year')
plt.ylabel('Average weekly sales')
plt.title('Trend of Average weekly sales over the years')

plt.grid(True)
plt.show()
```

```python
# by the above 2 graphs we can see that the average unemployment rate has been decreased over the years from 2010 to 2012 but also the average weekly sales also has been decreased over the years.
# we need to drill down by taking other factors as well.
```

```python
average_cpi_year = df.groupby('year').agg({'CPI': 'mean'})
average_cpi_year
# checking the consumer purchase index over the year to see where there is inflation or not which might impact the sales of the walmart
```

```python
plt.plot(average_cpi_year.index, average_cpi_year['CPI'], linestyle='-')

plt.xlabel('Year')
plt.ylabel('Average cpi ')
plt.title('Trend of consumer purchase index over the years')

plt.grid(True)
plt.show()
```

```python
#keynote- by the above graph we can see that consumer purchase index is increased year on year, increase in higher consumer index means higher inflation which indicates that cpi has been one of the factor influenced the sales of the walmart
```

```python
average_fuelprice_year = df.groupby('year').agg({'Fuel_Price': 'mean'})
average_fuelprice_year
# checking wheather the fuel prices has been increased or not over the years which might be a factor which influence the purchasing power among the consumers
```

```python
plt.plot(average_fuelprice_year.index, average_fuelprice_year['Fuel_Price'], linestyle='-')

plt.xlabel('Year')
plt.ylabel('average Fuel_Price  ')
plt.title('Trend of consumer fuel prices over the years')

plt.grid(True)
plt.show()
```

```python
# the fuel prices has drastically increased over the years which impact the cpi index which eventually effects the sales of the walmart
```
